# Reading from CSV File in Volume

```python
df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .csv('/Volumes/aerovista_bootcamp/bronze/source_systems/source_data/yellow_tripdata_2015-01.csv', header=True))

display(df)
```

# Write to Bronze Layer

```python
df.write.mode("overwrite").saveAsTable("aerovista_bootcamp.bronze.travel_trips")
```
